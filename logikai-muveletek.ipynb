# Logikai műveletek

A matematikai logika és a programozás IGAZ és HAMIS logikai értékekkel foglalkozik. Az ezekhez tartozó műveleteket fogom most röviden bemutatni.

````python
a = True
b = False
````

Egy pythonos kódblokkra kattintva, majd a felső sorban a ⏯️ Run gombra kattintva futtathatjátok a programrészletet. A kódblokkok összefüggenek, egy korábbi blokkban létrehozott változók egy későbbi blokkban is elérhetőek maradnak.

## Tagadás
A legelső és legegyszerűbb művelet a tagadás, ez igazból hamisat, hamisból igazat csinál. Szóban "nem"-ként mondjuk ki, matematikai jelölése &not;, pythonban a `not` kulcsszó, más programozási nyelvekben a `!` jelöli.

Legyen V egy logikai érték:

V | &not; V
--- | ---
Igaz | Hamis
Hamis | Igaz

Ebből az úgynevezett *igazságtáblából* látszik, hogy V milyen kiindulási értékéből milyen eredményt "csinál" a logikai művelet.

Kódban:

````python
print("`nem igaz` eredménye:", not True)
print("`nem hamis` eredménye:", not False)
````

## És
Az és művelettel két értékből "csinálunk" egyet. Ha mindkét oldala igaz, akkor az eredmény is igaz, különben hamis. Matematikai jelölése &and;, pyhthonban az `and` kulcsszó, más programozási nyelvekben a `&&` jelöli.

Legyen A és B két logikai érték:

A | B | A &and; B
---|---|---
Igaz|Igaz|Igaz
Igaz|Hamis|Hamis
Hamis|Igaz|Hamis
Hamis|Hamis|Hamis

````python
print("`igaz és igaz` eredménye:", True and True)
print("`igaz és hamis` eredménye:", True and False)
print("`hamis és igaz` eredménye:", False and True)
print("`hamis és hamis` eredménye:", False and False)
````

## Vagy
A vagy művelet szintén két értékből "csinál" egyet. Ha legalább az egyik oldala igaz, akkor az eredmény is igaz, ha mindkét oldala hamis, akkor az eredmény is hamis. Matematikai jelölése &or; pythonban az `or` kulcsszó, más programozási nyelvekben a `||` jelöli.

Legyen A és B két logikai érték:

A | B | A &or; B
---|---|---
Igaz|Igaz|Igaz
Igaz|Hamis|Igaz
Hamis|Igaz|Igaz
Hamis|Hamis|Hamis

````python
print("`igaz vagy igaz` eredménye:", True or True)
print("`igaz vagy hamis` eredménye:", True or False)
print("`hamis vagy igaz` eredménye:", False or True)
print("`hamis vagy hamis` eredménye:", False or False)
````

## Kizáró vagy
Vannak további logikai műveletek is, de nem fogjuk most mindegyiket megnézni. Egy "összetettebb" műveletet viszont igen, a kizáró vagyot: ez akkor igaz, ha CSAK az egyik oldala igaz. Általában `XOR` műveletként szoktuk írni, matematikai jelölése &veebar;; kódban viszont általában nincs rá külön művelet: vagy más logikai műveletekből vezetik le, vagy más műveletcsaládot használnak fel újra (ha valakit érdekel, hogy hogyan, a "Bitwise operátorok" témakörnek nézzen utána).

Legyen A és B két logikai érték:

A | B | A &veebar; B
---|---|---
Igaz|Igaz|Hamis
Igaz|Hamis|Igaz
Hamis|Igaz|Igaz
Hamis|Hamis|Hamis

Az `and` vagy `or` műveletekhez hasonló `xor` nincs pythonban, helyette egy okos következtetést tudunk használni: az *igazságtáblából* látszik, hogy csak akkor igaz a művelet eredménye, ha A és B különböző, vagy másképpen mondva: nem egyenlő.

````python
print("`igaz XOR igaz` eredménye:", True != True)
print("`igaz XOR hamis` eredménye:", True != False)
print("`hamis XOR igaz` eredménye:", False != True)
print("`hamis XOR hamis` eredménye:", False != False)
````

## Műveletek összeépítése
Ahogy a matematikai műveleteket, úgy a logikai műveleteket is össze lehet építeni. Ugyanúgy működik a zárójelezés, illetve a &not; művelet előnyt élvez a többivel szemben. Pl:

&not;(Igaz &or; Hamis) &and; (Igaz &and; Hamis)

```
nem (Igaz vagy Hamis) és (Igaz és Hamis)
nem (Igaz)            és (Hamis)
Hamis                 és (Hamis)
Hamis
```
